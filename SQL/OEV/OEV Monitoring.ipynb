```python
import pandas as pd

def process_calls(file_path, start_date_last, end_date_last, start_date, end_date):
    # Lire le fichier Excel
    appel = pd.read_excel(file_path)
    
    # Renommer les colonnes
    appel.rename(columns={'form.appels_oev.date_appel': 'date_appel',
                          'form.appels_oev.parenttuteur_trouve': 'found',
                          'form.appels_oev.patient_code': 'patient_code'}, inplace=True)
    
    # Convertir les colonnes en datetime et numérique
    appel['date_appel'] = pd.to_datetime(appel['date_appel'], errors='coerce')
    appel['found'] = pd.to_numeric(appel['found'], errors='coerce')
    
    # Filtrer les données pour les périodes spécifiées
    appel_qi = appel[(appel['date_appel'] >= start_date_last) & (appel['date_appel'] <= end_date_last) &
                     (appel['found'] == 1)][['patient_code', 'date_appel', 'found']].drop_duplicates('patient_code')
    appel_qj = appel[(appel['date_appel'] >= start_date) & (appel['date_appel'] <= end_date) &
                     (appel['found'] == 1)][['patient_code', 'date_appel', 'found']].drop_duplicates('patient_code')
    
    # Créer une table similaire à la table des services
    all_appel_served_q1q2 = appel_qi.merge(appel_qj[['patient_code', 'date_appel']], on='patient_code', how='left')
    
    # Calculer les résultats
    nombre_appel_qi = appel_qi.shape[0]
    nombre_appel_qj = appel_qj.shape[0]
    servis_q1_q2 = all_appel_served_q1q2[all_appel_served_q1q2['date_appel_y'].notnull()].shape[0]
    
    return nombre_appel_qi, nombre_appel_qj, servis_q1_q2

# Exemple d'utilisation
file_path = "C:\\Users\\Moise\\Downloads\\Reports_MEAL\\OEV\\Caris Health Agent - Enfant - APPELS OEV (created 2022-03-28) 2025-03-04.xlsx"


nombre_appel_qi, nombre_appel_qj, servis_q1_q2 = process_calls(file_path, start_date_last, end_date_last, start_date, end_date)

print(f"Nombre d'appels Qi: {nombre_appel_qi}")
print(f"Nombre d'appels Qj: {nombre_appel_qj}")
print(f"Servis en Q1 et Q2: {servis_q1_q2}")
```

```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
from dateutil.parser import parse

# Ensure necessary variables are defined

# Dates for the most recent period
start_date = pd.to_datetime('2025-01-01')
today_date = datetime.today()
end_date = today_date

# Dates for the period before the most recent period
start_date_last = pd.to_datetime('2024-10-01')
end_date_last = pd.to_datetime('2024-12-31')

# Creating the age column
def get_age(date):
    if pd.isnull(date):
        return None
    return relativedelta(end_date, parse(str(date))).years

# Convert 'date_of_birth' column to datetime, with errors='coerce' to handle invalid dates
oev_data_on_hivhaiti['date_of_birth'] = pd.to_datetime(oev_data_on_hivhaiti['date_of_birth'], errors='coerce')

# Map the 'get_age' function to the 'date_of_birth' column
oev_data_on_hivhaiti['ovc_age'] = oev_data_on_hivhaiti['date_of_birth'].map(get_age)


display(oev_data_on_hivhaiti[['date_of_birth','ovc_age']].head(2))

```

```python
def process_age_ranges(df, age_column, separation_age):

    # Convertir la colonne d'âge en numérique
    df[age_column] = pd.to_numeric(df[age_column], errors='coerce')

    # Fonction pour obtenir la tranche d'âge
    def age_range(age):
        if pd.isna(age):
            return None
        elif age in range(0, 1):
            return '< 1'
        elif age in range(1, 5):
            return '1 - 4'
        elif age in range(5, 10):
            return '5 - 9'
        elif age in range(10, 15):
            return '10 - 14'
        elif age in range(15, 18):
            return '15 - 17'
        elif age in range(18, 21):
            return '18 - 20'
        else:
            return '21+'

    # Ajouter une colonne 'age_range' au DataFrame
    df['age_range'] = df[age_column].map(age_range)

    # Filtrer les données en fonction des tranches d'âge
    df1 = df[df[age_column] < separation_age]
    df2 = df[(df[age_column] >= separation_age) & (df[age_column] <= 24)]

    # Afficher le nombre d'observations dans chaque groupe
    print(f"Ce dataset contient {df1.shape[0]} observations pour les tranches d'âge inférieures à {separation_age}.")

    print(f"Ce dataset contient {df2.shape[0]} observations pour les tranches d'âge compris entre {separation_age} et 24 ans.")
    # Afficher les résultats
    print(f"DataFrame pour les âges inférieurs à {separation_age}:")
    display(df1.head(2))
    print(f"DataFrame pour les âges compris entre {separation_age} et 24 ans:")
    display(df2.head(2))

    # Retourner les deux DataFrames
    return df1, df2
```

```python
def active_df(df):
    # Départements à exclure
    excluded_departments = ['Sud', "Grand'Anse", 'Nippes']

    # Filtrer les données selon les critères.  Handle potential missing values gracefully.
    df = df[
        (df['is_abandoned'] != 1) &
        (df['is_dead'] != 1) &
        (df['is_ugp'] == 'No') &
        (df['network_x'] != 'UGP') &
        (df['network_x']!= 'PIH') &
        (df['LTFU 30days'] =="No") &
        (~df['Departement'].isin(excluded_departments))  # Exclure les départements
    ]

    return df
```
